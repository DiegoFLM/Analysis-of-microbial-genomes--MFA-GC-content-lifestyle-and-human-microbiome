```python
def get_largest_fna_files(main_dir: Path):
    largest_sequences = []
    # Ensure the provided path is a directory
    if not main_dir.is_dir():
        print(f"The path {main_dir} is not a directory.")
        return
    # Iterate through each subdirectory in the main directory
    for sub_dir in main_dir.iterdir():
        if not sub_dir.is_dir():
            continue  # Skip if it's not a directory
        # print(f"Processing subdirectory: {sub_dir}")

        # Find all .fna files in the subdirectory
        fna_files = list(sub_dir.glob("*.fna"))
        if not fna_files:
            print(f"No .fna files found in {sub_dir}. Skipping...")
            continue

        # Identify the largest .fna file
        largest_file = max(fna_files, key=lambda f: f.stat().st_size)
        largest_sequences.append(largest_file)

        # print(f"Largest file: {largest_file} ({largest_file.stat().st_size} bytes)")

    return largest_sequences
```

```python
paths_largest_seqs_hm = get_largest_fna_files(PATH_HM_FNA)
print(len(paths_largest_seqs_hm))

for element in paths_largest_seqs_hm[:5]:
    print(element)
```

```python
destiny_path_archaea = PATH_RESULTS_ARCHAEA / 'mfa_archaea_pow7.csv'
```

```python
df_archaea = pd.read_csv(PATH_DATA / 'raw' / \
    'scraped' / 'archaea_name_url.csv')
df_archaea
```

```python
df_mfa_archaea = pd.read_csv(destiny_path_archaea, sep=';')
df_mfa_archaea
```

```python
df_mfa_url_archaea = df_mfa_archaea.copy()
```

```python
df_archaea['Organism'] = df_archaea['name'].str.lower()
df_archaea
```

```python
# Function to create new rows for each organism
def generate_segments(row):
    # Original row
    original_row = row.copy()
    
    # Create segment rows
    segment_1 = row.copy()
    segment_1['Organism'] = row['Organism'] + '_segment_1'
    
    segment_2 = row.copy()
    segment_2['Organism'] = row['Organism'] + '_segment_2'
    
    segment_3 = row.copy()
    segment_3['Organism'] = row['Organism'] + '_segment_3'
    
    # Combine all rows into a list
    return [original_row, segment_1, segment_2, segment_3]
```

```python
df_temp_archaea = df_archaea.apply(lambda row: pd.DataFrame(generate_segments(row)), axis=1)
df_archaea = pd.concat(df_temp_archaea.values)
df_archaea.reset_index(drop=True, inplace=True)
df_archaea
```

```python
df_mfa_url_archaea = df_mfa_url_archaea.merge(df_archaea, on='Organism', how='left')
df_mfa_url_archaea = df_mfa_url_archaea[ ['name', 'url' , 'Organism', 'path', 'seq_length', 'GC_content', 'Q', 'Tau(Q)', 'D(Q)', 'r_squared', 'Delta_Dq'] ]
df_mfa_url_archaea
```
